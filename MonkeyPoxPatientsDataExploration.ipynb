```python
#DATA V ISUALISATIONS
#percentage for each category
category_value = Mehak_df['MonkeyPox'].value_counts()
labels_category=category_value.index
explode = (0,0.1)
colors = ('lavenderblush','plum')
plt.pie(category_value, labels=labels_category, explode = explode, colors = colors, autopct='%1.1f%%', shadow=True)
plt.title("Percentage of Negative and Positive Cases");
```

```python
MonkeyPox_Labels = Mehak_df.MonkeyPox.value_counts()
print(MonkeyPox_Labels);
```

```python
#number of positive and negative cases
sns.barplot(x=MonkeyPox_Labels.index, y=MonkeyPox_Labels, palette = 'RdPu_r')
plt.xlabel('Result')
plt.ylabel('Number')
plt.title('Number of negative and positive cases');
```
